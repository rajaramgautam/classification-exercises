# 1. Attendance Data

# Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

# You should end up with something like this:


# name
# Billy    0.5250
# Jane     0.6875
# John     0.9125
# Sally    0.7625
# Name: grade, dtype: float64

```python
import pandas as pd
import numpy as np
import seaborn as sns
from pydataset import data
```

```python
attendance = pd.read_csv('untidy-data/attendance.csv')
attendance
```

```python
attendance.info()
```

```python
#rename columns
attendance.columns
attendance.columns = ['name', '2018-01-01', '2018-01-02', '2018-01-03', '2018-01-04', '2018-01-05', '2018-01-06', '2018-01-07', '2018-01-08']
attendance
```

```python
# Melting data except name
attendance= attendance.melt(id_vars = 'name', var_name = 'date', value_name = 'roll_call')
attendance
```

```python
def calc_point(roll_call):
    if roll_call == "P":
        point = 1
    elif roll_call == "H":
        point = 0.5
    elif roll_call == "T":
        point = 0.9
    else:
        point = 0
        
    return point
        
```

```python
attendance.info()
```

```python
attendance['attendance_point'] = attendance.roll_call.apply(calc_point)
attendance
```

```python
attendance.groupby(['name']).mean()
```

# ALTERNATIVE METHOD:

```python
attendance['point'] = attendance.roll_call.replace({'P': 1, 'A': 0, 'H': 0.5, 'T': 0.9})
attendance
```

```python
attendance.groupby(['name']).mean()
```

# 3. Cake Recipes

- a. Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
cake_recipes = pd.read_csv('untidy-data/cake_recipes.csv')
cake_recipes.head()
```

```python
cake_recipes.info()
```

- b. Tidy the data as necessary.

```python
cake_recipes[['recipe', 'position']] =cake_recipes['recipe:position'].str.split(':', expand = True)
cake_recipes.head()
```

```python
cake_recipes = cake_recipes.drop(columns = 'recipe:position')
cake_recipes
```

```python
cake_recipes_melt = cake_recipes.melt(id_vars = ['recipe', 'position'], var_name = 'temperature', value_name = 'score')
cake_recipes_melt.head(6)
```

- c. Which recipe, on average, is the best? recipe b

```python
cake_recipes_melt.groupby(['recipe']).score.mean()
```

- d. Which oven temperature, on average, produces the best results? 275

```python
cake_recipes_melt.groupby('temperature').score.mean()
```

- e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_recipes_melt.groupby(['recipe', 'position', 'temperature']).score.mean()
```
